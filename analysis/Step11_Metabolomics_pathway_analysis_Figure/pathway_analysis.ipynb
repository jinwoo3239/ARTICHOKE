```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'Arial'

import warnings
warnings.filterwarnings("ignore")
```

```python
# The pathway.csv dataset was generated using MetaboAnalyst based on 92 key metabolites identified from the analysis.


df = pd.read_csv("./data/pathway.csv")
df_bubble = df.sort_values('-log(p)', ascending=True)
```

```python
plt.figure(figsize=(6, 2.6))

scatter = plt.scatter(
    x=df_bubble['-log(p)'],
    y=df_bubble['Pathway Name'],
    s=df_bubble['Hit ratio'] * 400,   #   
    color='gray',
    alpha=0.8,
    edgecolors='black'
)

for ratio in [0.1, 0.2, 0.3]:
    plt.scatter([], [], s=ratio * 400, color='gray', edgecolors='black', label=f'{ratio:.1f}')

plt.legend(
    title='Hit ratio',
    scatterpoints=1,
    bbox_to_anchor=(1.05, 0.5),
    loc='center left',
    frameon=True,
    labelspacing=1.2,
    borderpad=1,
    fontsize=9
)

plt.xlim(0, 5.5)

plt.xticks([0, 1, 2, 3, 4, 5],fontsize=9)
plt.yticks(fontsize=9)
plt.xlabel('-log(p-value)', fontsize=9)
plt.ylabel('Pathway', fontsize=9)

plt.tight_layout()
plt.savefig('pathway.tif', dpi=600)
plt.show()
```
